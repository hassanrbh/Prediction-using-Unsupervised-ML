## Finding the optimal number of clusters for K-Means and determining the value of K :

```python
x = task_2.iloc[: , [1 , 2 , 3 ,4]].values

# Set Up Kmean Algorithm :
from sklearn.cluster import KMeans
sum_of_the_square_distance = []
range_ = range(1 , 13)
for i in range_:
    km = KMeans(n_clusters = i , max_iter = 350, n_init = 12, random_state = 0)
    km = km.fit(x)
    sum_of_the_square_distance.append(km.inertia_)

# PLot The Optimal Number of Clusters :
plt.plot(range_ , sum_of_the_square_distance , 'bx-')
plt.xlabel('k')
plt.ylabel('Sum of The Square Distance')
plt.title('Optinal Number of Clusters')
plt.show()


# Answer : In This PLot Above The Elbow at k = 3 indicating the optimal k for this dataset is 3.
```

# Creating K-Means Classifier :

```python
kmeans = KMeans(n_clusters=3 , max_iter = 350, n_init = 12, random_state = 0)
kmeans.fit(x)
y_means = kmeans.predict(x)
```

### Let's visualize the results by plotting the data colored by these labels. 
### We will also plot the cluster centers as determined by the k-means estimator:

```python
plt.scatter(x[:, 0], x[:, 1], c=y_means, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
